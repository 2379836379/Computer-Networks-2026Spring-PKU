# scapy安装简介

# scapy案例

准备环境

```python
from scapy.all import *
```

## 嗅探报文

```python
packets = sniff(timeout=15)  # 5秒内在"外一个终端窗口"命令行运行 wget http://baidu.com
print(len(packets))
```

## 存储报文至文件

```python
wrpcap("tutorial.pcap", packets)
```

## 从文件读取报文

```python
packets = rdpcap("tutorial.pcap")
```

## 显示报文（自动解析结果）

```python
for pkt in packets[-2:]:
    pkt.show()
```

## 查看报文类的成员

```python
p1 = packets[-1]
print(p1)
```

## 逐层解析报文

```python
print(p1.type)
```

2048是0x800，因此是IP分组

```python
p2 = p1.payload
print(p2)
```

```python
print(p2.proto)
```

6表示TCP，17表示UDP

```python
p3=p2.payload
print(p3)
```

## 可视化报文

生成一个可视化的pdf文件。下面命令可能失败，缺少依赖包。如果运行失败，跳过本步。

```python
p1.pdfdump('pkt.pdf')
```

## 其他解析报文的方法

IP、TCP、UDP、Ether是scapy内部定义的协议类型

```python
print('IP' in p1)
print('TCP' in p1)
print('UDP' in p1)
print('Ether' in p1)
```

使用haslayer方法

```python
print(p1.haslayer(Ether))
print(p1.haslayer(IP))
print(p1.haslayer(TCP))
print(p1.haslayer(UDP))
```
